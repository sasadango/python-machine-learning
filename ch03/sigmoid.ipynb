```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# define sigmoid
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
```

```python
# interval:0.1 range(-7, 7)
z = np.arange(-7, 7, 0.1)

# do
phi_z = sigmoid(z)

# plot
plt.plot(z, phi_z)
plt.axvline(0.0, color='k')
plt.ylim(-0.1, 1.1)
plt.xlabel('z')
plt.ylabel('$\phi (x)$')
plt.yticks([0.0, 0.5, 1.0])
ax = plt.gca()
ax.yaxis.grid(True)
plt.tight_layout()
plt.show()
```
